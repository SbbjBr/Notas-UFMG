### Transformar respectivos valores

```python
def retornar_nota(nota):
    
    """
    Substitui a vírgula por um ponto para se adequar ao sistema
    númerico internacional e retorna a nota em float
    """
    
    nota = nota.replace(",",".")
    return float(nota)
```

```python
def corrigir_modalidade(modalidade):
    
    """
    Corrige o erro da modalidade 'Ampla concorrência' estar escrita errado ou
    adiciona o nome dela
    "Cor, Renda, Escola P", "Renda, Escola", "Cor, Escola", "Escola P", "Ampla concorrência"
    
    """
    
    if modalidade == "Ampla\rconcorrência":
        return "Ampla concorrência"
    
    elif modalidade == "1.2":
        return "Cor, Renda, Escola"
    
    elif modalidade == "2.2":
        return "Renda, Escola"
    
    elif modalidade == "3.2":
        return "Cor, Escola"
    
    elif modalidade == "4.2":
        return "Escola"
    else:
        return modalidade
    
```
